### 1. AC

```python
import sys
from io import StringIO
from collections import deque

sys.stdin=StringIO("""\
4
RDD
4
[1,2,3,4]
DD
1
[42]
RRD
6
[1,1,2,3,5,8]
D
0
[]
""")
input=sys.stdin.readline

t = int(input())
for _ in range(t):
    p = input()
    n = int(input())
    nums = input().strip()
    if n==0:
        nums = deque([])
    else:
        nums = deque(nums[1:-1].split(','))
    error = False
    is_reversed = False
    for one_p in p:
        if one_p == 'R':
            is_reversed = not is_reversed
        elif one_p == 'D':
            if len(nums) <1:
                print('error')
                error = True
                break
            if is_reversed:
                nums.pop()
            else:
                nums.popleft()
    if is_reversed:
        nums.reverse()

    if not error:
        print("[" + ",".join(nums) + "]")
```

### 2. 시리얼 번호

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
4
34H2BJS6N
PIM12MD7RCOLWW09
PYF1J14TF
FIPJOTEA5
""")
input=sys.stdin.readline

result = []
n = int(input())
for _ in range(n):
    num = input().strip()
    sum_n = 0
    for n in num:
        if n.isdigit():
            sum_n += int(n)
    result.append((len(num), sum_n, num))

result.sort()
for _,_,num in result:
    print(num)
```

### 3. DNA 비밀번호

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
4 2
GATA
1 0 0 1
""")
input=sys.stdin.readline

s, p = map(int, input().split())
string = input()
min_nums = list(map(int, input().split()))
curr = {'A':0, 'C':0, 'G':0, 'T':0}

result = 0

def check():
    for i in range(4):
        if curr['ACGT'[i]] < min_nums[i]:
            return False
    return True

for i in range(p):
    curr[string[i]] += 1

if check():
    result+=1

for i in range(p,s):
    curr[string[i]]+=1
    curr[string[i-p]]-=1
    
    if check():
        result+=1

print(result)
```

### 4. 영단어 암기는 괴로워

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
12 5
appearance
append
attendance
swim
swift
swift
swift
mouse
wallet
mouse
ice
age
""")
input=sys.stdin.readline

n, m = map(int, input().split())
words = {}
for _ in range(n):
    word = input().strip()
    words[word] = words.get(word, 0)+1

tosort=[]
for word in words.keys():
    if len(word) >= m:
        tosort.append((-words[word], -len(word), word))

tosort.sort()

for _,_,word in tosort:
    print(word)
```

### 5. 단어 뒤집기 2

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
<problem>17413<is hardest>problem ever<end>
""")

input=sys.stdin.readline

stack = []

S = input().strip()+ ' '
is_tag = False

for s in S:
    if s=='<':
        is_tag=True
        if stack:
            while stack:
                print(stack.pop(), end='')
        print(s, end='')
        
    elif s=='>':
        is_tag=False
        print(s, end='')
    
    elif is_tag==True:
        print(s, end='')
    
    elif is_tag==False:
        if s!=' ':
            stack.append(s)
        else:
            if stack:
                while stack:
                    print(stack.pop(), end='')
            print(s, end='')
```

### 6. 서로 다른 부분 문자열의 개수

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
ababc
""")

input=sys.stdin.readline

S = input().strip()

result = []

for length in range(len(S)):
    for i in range(len(S)-length):
        result.append(S[i:i+length+1])

print(len(set(result)))
```

### 7. IOIOI

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
2
13
OOIOIOIOIIOII
""")

input=sys.stdin.readline

n = int(input())
m = int(input())
S = input().strip()

p1 = 'IOI'
pn = p1+'OI'*(n-1)

result = 0

for i in range(m-len(pn)+1):
    # print(S[i:i+len(pn)])
    if S[i:i+len(pn)] == pn:
        result+=1
        
print(result)
```

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
2
13
OOIOIOIOIIOII
""")

input=sys.stdin.readline

n = int(input())
m = int(input())
S = input().strip()

result = 0
cnt=0
i=0

while i<(m-2):
    if S[i]+S[i+1]+S[i+2]=='IOI':
        cnt+=1
        if cnt==n:
            result+=1
            cnt-=1
        i+=2
    else:
        i+=1
        cnt=0

print(result)
```

### 8. A와 B

```python
import sys
from io import StringIO
from collections import deque

sys.stdin=StringIO("""\
AB
ABB
""")

input=sys.stdin.readline

S = input().strip()
T = input().strip()
t_a = T.count('A')
t_b = T.count('B')

ss = deque([S])

cant=False

while ss:
    string = ss.popleft()
    if string==T:
        print(1)
        cant = True
        break
    if len(string) < len(T) and (string.count('A')<t_a or string.count('B')<t_b):
        ss.append(string+'A')
        ss.append(string[::-1]+'B')

if not cant:      
    print(0)
```

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
B
ABBA
""")

input=sys.stdin.readline

S = input().strip()
T = input().strip()

while len(T)>len(S):
    if T[-1]=='A':
        T = T[:-1]
    elif T[-1]=='B':
        T = T[:-1][::-1]
        
if T==S:
    print(1)
else:
    print(0)
```

### 9. JadenCase 문자열 만들기

```python
def solution(s):
    answer = []
    ss = s.split(" ")
    for word in ss:
        if word =="":
            answer.append("")
            continue
        if word[0].isalpha():
            after = word[0].upper()+word[1:].lower()
        else:
            after = word[0]+word[1:].lower()
        answer.append(after)
    return ' '.join(answer)

s = "  for the what  1what  "
solution(s)
```
