### 1. AC

```python
import sys
from io import StringIO
from collections import deque

sys.stdin=StringIO("""\
4
RDD
4
[1,2,3,4]
DD
1
[42]
RRD
6
[1,1,2,3,5,8]
D
0
[]
""")
input=sys.stdin.readline

t = int(input())
for _ in range(t):
    p = input()
    n = int(input())
    nums = input().strip()
    if n==0:
        nums = deque([])
    else:
        nums = deque(nums[1:-1].split(','))
    error = False
    is_reversed = False
    for one_p in p:
        if one_p == 'R':
            is_reversed = not is_reversed
        elif one_p == 'D':
            if len(nums) <1:
                print('error')
                error = True
                break
            if is_reversed:
                nums.pop()
            else:
                nums.popleft()
    if is_reversed:
        nums.reverse()

    if not error:
        print("[" + ",".join(nums) + "]")
```

### 2. 시리얼 번호

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
4
34H2BJS6N
PIM12MD7RCOLWW09
PYF1J14TF
FIPJOTEA5
""")
input=sys.stdin.readline

result = []
n = int(input())
for _ in range(n):
    num = input().strip()
    sum_n = 0
    for n in num:
        if n.isdigit():
            sum_n += int(n)
    result.append((len(num), sum_n, num))

result.sort()
for _,_,num in result:
    print(num)
```

### 3. DNA 비밀번호

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
4 2
GATA
1 0 0 1
""")
input=sys.stdin.readline

s, p = map(int, input().split())
string = input()
min_nums = list(map(int, input().split()))
curr = {'A':0, 'C':0, 'G':0, 'T':0}

result = 0

def check():
    for i in range(4):
        if curr['ACGT'[i]] < min_nums[i]:
            return False
    return True

for i in range(p):
    curr[string[i]] += 1

if check():
    result+=1

for i in range(p,s):
    curr[string[i]]+=1
    curr[string[i-p]]-=1
    
    if check():
        result+=1

print(result)
```

### 4. 영단어 암기는 괴로워

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
12 5
appearance
append
attendance
swim
swift
swift
swift
mouse
wallet
mouse
ice
age
""")
input=sys.stdin.readline

n, m = map(int, input().split())
words = {}
for _ in range(n):
    word = input().strip()
    words[word] = words.get(word, 0)+1

tosort=[]
for word in words.keys():
    if len(word) >= m:
        tosort.append((-words[word], -len(word), word))

tosort.sort()

for _,_,word in tosort:
    print(word)
```

### 8. A와 B

```python
import sys
from io import StringIO
from collections import deque

sys.stdin=StringIO("""\
AB
ABB
""")

input=sys.stdin.readline

S = input().strip()
T = input().strip()
t_a = T.count('A')
t_b = T.count('B')

ss = deque([S])

cant=False

while ss:
    string = ss.popleft()
    if string==T:
        print(1)
        cant = True
        break
    if len(string) < len(T) and (string.count('A')<t_a or string.count('B')<t_b):
        ss.append(string+'A')
        ss.append(string[::-1]+'B')

if not cant:      
    print(0)
```

### 10. 올바른 괄호

```python
def solution(s):
    stack=[]
    for ss in s:
        if ss=='(':
            stack.append(ss)
        else:
            if not stack:
                return False
            else:
                stack.pop()
    if stack:
        return False
    return True
```

### 13. 숫자의 표현

```python
from collections import deque

def solution(n):
    answer = 0
    now=deque([1])
    i, j=1,1
    while i<=j:
        if sum(now)==n:
            answer+=1
            i+=1
            now.popleft()
        elif sum(now)<n:
            j+=1
            now.append(j)
        else:
            i+=1
            now.popleft()
    return answer
```

### 16. 짝지어 제거하기

```python
#효율성 주의** -> stack 쓰기
def solution(s):
    stack=[]
    for ss in s:
        if stack and stack[-1]==ss:
            stack.pop()
        else:
            stack.append(ss)
    return 0 if stack else 1
```

### 18. 조이스틱

```python
def solution(name):
    #알파벳 변경 횟수
    spell_cnt = 0
    for char in name:
        spell_cnt+=min(ord(char)-ord('A'), ord('Z')-ord(char)+1)
        
    #좌우 이동 횟수
    n = len(name)
    move_cnt= n-1
    for i in range(n):
        nxt = i+1
        while nxt<n and name[nxt]=='A':
            nxt+=1
        move_cnt = min(move_cnt, 2*i+(n-nxt), 2*(n-nxt)+i)
    return spell_cnt+move_cnt
```

### 19. 큰 수 만들기

```python
def solution(number, k):
    stack = []
    for num in number:
        while k>0 and stack:
            if stack[-1]<num:
                stack.pop()
                k-=1
            else:
                break
        stack.append(num)
    if len(stack)> (len(number)-k):
        stack = stack[:len(number)-k]
    return "".join(stack)
```

### 20. 구명보트

```python
def solution(people, limit):
    answer = 0
    people.sort()
    start, end = 0, len(people)-1
    while start<=end:
        if (people[start]+people[end]) <= limit:
            start+=1
        end-=1  #가벼운 사람 못태워도 일단 무거운 사람은 보내기      
        answer+=1
    return answer
```

### 21. 귤고르기

```python
def solution(k, tangerine):
    answer = 0
    cnt_tan = {}
    for num in tangerine:
        cnt_tan[num] = cnt_tan.get(num, 0)+1
        
    cnts = sorted(cnt_tan.values(), reverse=True)
    
    for cnt in cnts:
        k-=cnt
        answer+=1
        if k<=0:
            break
    return answer
```

### 22. 멀리 뛰기

```python
def solution(n):
    answer = 0
    dp = [0]*(n+1)
    dp[1]=1
    if n >1:
        dp[2]=2
        for i in range(3, n+1):
            dp[i] = dp[i-1]+dp[i-2]
    return dp[n]%1234567
```

### 23. N개의 최소공배수

```python
def solution(arr):
    arr.sort()
    i=1
    while True:
        all = True
        for j in range(len(arr)-1):
            if (arr[-1]*i)%arr[j]!=0:
                i+=1
                all = False
                break
        if all == False:
            continue
        else:
            answer = arr[-1]*i
            break
    return answer
```

### 24. 연속 부분 수열 합의 개수

```python
def solution(elements):
    n = len(elements)
    ssum = set()
    for i in range(n):
        s = elements[i]
        ssum.add(s)
        for j in range(i+1, i+n):
            s+=elements[j%n]
            ssum.add(s)
    return len(ssum)
```

### 막혔던 문제들 다시 풀기

```python
#DNA 비밀번호
import sys
from io import StringIO

sys.stdin=StringIO("""\
4 2
GATA
1 0 0 1
""")

input=sys.stdin.readline

s, p = map(int, input().split())
string = input()
min_str = list(map(int, input().split())) #ACGT
min_dict = {'A': min_str[0], 'C': min_str[1], 'G':min_str[2], 'T':min_str[3]}
cnt = {'A':0, 'C':0, 'G':0, 'T':0}

answer = 0

for i in range(s-p+1):
    can = True
    if i ==0:
        for j in range(i, i+p):
            cnt[string[j]]+=1
    else:
        cnt[string[i-1]]-=1
        cnt[string[i+p-1]]+=1
    
    for chr in cnt.keys():
        if cnt[chr]< min_dict[chr]:
            can = False
            break
    
    if can:
        answer+=1
        
print(answer)
```
