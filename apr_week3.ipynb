### 1. A → B

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
100 40021
""")

input=sys.stdin.readline

a, b = map(str, input().split())

oper = 0
while int(b)>1 and int(a)!=int(b):
    if b[-1] == "1":
        b = b[:-1]
    elif int(b)%2==0:
        b = str(int(b)//2)
    else:
        break
    oper+=1

if int(b) != int(a):
    print(-1)
else:
    print(oper+1)
```

### 2. 부등호

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
9
> < < < > > > < <
""")

input=sys.stdin.readline

k = int(input())
k_list = list(input().split())

visited = [False]*(10)
result = []

def check(a, b, op):
    if op == '>':
        return a>b
    if op == '<':
        return a<b
    
def dfs(depth, s):
    if depth==k+1:
        result.append(s)
        return 
    
    for i in range(10):
        if visited[i]==False:
            if depth==0 or check(int(s[-1]), i, k_list[depth-1]):
                visited[i]=True
                dfs(depth+1, s+str(i))
                visited[i]=False

dfs(0, '')
print(max(result))
print(min(result))
```
