### 2. 알고리즘 수업 - 병합 정렬 1

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
5 7
4 5 1 3 2
""")
input=sys.stdin.readline

N, K = map(int, input().split())
a = list(map(int, input().split()))

def merge_sort(A,p,r,K):
    if p >= r:
        return 0, None  # (저장 횟수, K번째 저장값)

    q = (p+r)//2
    cnt_left, val = merge_sort(A, p, q, K)
    if val is not None:
        return cnt_left, val

    cnt_right, val = merge_sort(A, q + 1, r, K - cnt_left)
    if val is not None:
        return cnt_left + cnt_right, val

    cnt_merge, val = merge(A, p, q, r, K - (cnt_left + cnt_right))
    return cnt_left + cnt_right + cnt_merge, val

def merge(A, p, q, r, K):
    # A[p..q]와 A[q+1..r]은 이미 정렬되어 있다고 가정하고 병합
    tmp = []
    i, j = p, q + 1  #왼쪽, 오른쪽 시작점

    # 1) 작은 값부터 tmp에 담기 (여긴 "저장 카운트" 아님!)
    while i <= q and j <= r:
        if A[i] <= A[j]:
            tmp.append(A[i]); i += 1
        else:
            tmp.append(A[j]); j += 1

    while i <= q:
        tmp.append(A[i]); i += 1
    while j <= r:
        tmp.append(A[j]); j += 1

    # 2) tmp를 A[p..r]에 다시 복사  ← ★ 문제에서 말하는 "저장"이 여기!
    cnt = 0
    for idx in range(p, r + 1):
        A[idx] = tmp[idx - p]   # <- 이 한 줄이 "저장"
        cnt += 1                # 저장 횟수 1 증가
        if cnt == K:
            return cnt, A[idx]  # K번째 저장되는 값

    return cnt, None


# 예제 실행
_, ans = merge_sort(a, 0, len(a) - 1, K)
print(ans if ans is not None else -1)
```

### 3. N과 M (5)

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
4 4
1231 1232 1233 1234
""")
input=sys.stdin.readline

n, m = map(int, input().split())
nums = list(map(int, input().split()))
nums.sort()
visited=[False]*n
tmp=[]

def dfs():
    if len(tmp)==m:
        print(*tmp)
        return
        
    for i in range(n):
        if visited[i]==False:
            tmp.append(nums[i]) 
            visited[i]=True  
            dfs()
            tmp.pop()
            visited[i]=False
        
dfs()
```

### 4. 연산자 끼워넣기

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
6
1 2 3 4 5 6
2 1 1 1
""")
input=sys.stdin.readline

n = int(input())
nums = list(map(int, input().split()))
opers = list(map(int, input().split())) #덧셈, 뺄셈, 곱셈, 나눗셈
answers=[]

def dfs(idx, curr):
    if idx ==(n-1):
        answers.append(curr)
        return
        
    for op in range(4):
        if opers[op]>0:
            opers[op]-=1 
            if op == 0:
                new = curr+nums[idx+1]
            elif op ==1:
                new = curr-nums[idx+1]
            elif op==2:
                new = curr*nums[idx+1] 
            else:
                if curr>0:
                    new=curr//nums[idx+1]
                else:
                    new=-(-curr//nums[idx+1]) 
            dfs(idx+1, new) 
            opers[op]+=1
        
dfs(0, nums[0])
print(max(answers))
print(min(answers))
```
