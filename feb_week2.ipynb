### 1. 적록색약 : https://www.acmicpc.net/problem/10026

```python
from collections import deque
import sys
from io import StringIO

sys.stdin=StringIO("""\
5
RRRBB
GGBBB
BBBRR
BBRRR
RRRRR
""")

input = sys.stdin.readline

n = int(input())
graph=[]
graph2=[]
for i in range(n):
    pic = input()
    graph.append(list(pic.strip()))
    graph2.append(list(pic.replace('G','R').strip()))


def bfs(m, i,j):
    q=deque()
    q.append((i,j))
    col = m[i][j]
    m[i][j]=None
    while q:
        x, y = q.popleft()
        for dx, dy in [(-1,0),(1,0),(0,-1),(0,1)]:
            nx, ny = x+dx, y+dy
            if 0<=nx<n and 0<=ny<n and m[nx][ny]==col:
                q.append((nx, ny))
                m[nx][ny]=None

cnt=0
cnt2=0
for i in range(n):
    for j in range(n):
        if graph[i][j]:
            bfs(graph, i,j)
            cnt+=1
        if graph2[i][j]:
            bfs(graph2,i,j)
            cnt2+=1

print(cnt, cnt2)


#적록색약 : 빨강==초록
```

### 2. 이분그래프 : https://www.acmicpc.net/problem/1707

```python
import sys
from io import StringIO
from collections import deque
sys.stdin=StringIO("""\
2
3 2
1 3
2 3
4 3
1 2
2 3
3 4
""")

input=sys.stdin.readline

k = int(input())
for _ in range(k):
    V, E = map(int, input().split())  #정점개수 간선개수
    graph=[[]for _ in range(V+1)]
    color=[0]*(V+1)  #1 or -1로 색칠

    for _ in range(E):
        u, v = map(int, input().split())
        graph[u].append(v)
        graph[v].append(u)
    
    ok=True
    for stt in range(1, V+1):
        if color[stt]!=0:  #색칠 안된 정점에 대해서만 검사 
            continue
        color[stt]=1
        q=deque()
        q.append(stt)
        while q and ok:  #연결된 모든 정점들의 색은 다른색이어야함  
            x=q.popleft()
            for nxt in graph[x]:
                if color[nxt]==0: #아직 안칠해진 경우
                    color[nxt]=-color[x]
                    q.append(nxt)
                elif color[nxt]==color[x]: #같은색인 경우
                    ok=False
                    break
        if not ok:
            break
        

    print('YES' if ok else 'NO')
```

### 3.빙산

```python
import sys
from io import StringIO
from collections import deque
sys.stdin=StringIO("""\
5 7
0 0 0 0 0 0 0
0 2 4 5 3 0 0
0 3 0 2 5 2 0
0 7 6 2 4 0 0
0 0 0 0 0 0 0
""")
input=sys.stdin.readline
n,m = map(int, input().split())
sea=[]
for _ in range(n):
    sea.append(list(map(int, input().split())))
    
def bfs(i,j):
    q = deque()
    q.append((i,j))
    visited[i][j]=1
    while q:
        x, y = q.popleft()
        for dx, dy in [(-1,0),(1,0),(0,1),(0,-1)]:
            nx, ny=x+dx, y+dy
            if 0<=nx<n and 0<=ny<m:  
                if sea[nx][ny]==0:  #주변이 바다인 경우
                    visited[x][y]+=1
                if sea[nx][ny]!=0 and visited[nx][ny]==0:  #주변이 방문하지 않은 빙산인 경우
                    q.append((nx,ny))
                    visited[nx][ny]=1

year=0                 
while True:
    visited=[[0]*m for _ in range(n)]
    cnt=0  #덩어리 개수
    
    #덩어리 개수 세면서 visited에 바다 개수 업데이트
    for i in range(n):
        for j in range(m):
            if sea[i][j] > 0 and visited[i][j] == 0:
                bfs(i,j)
                cnt+=1
                
    if cnt>=2:
        print(year)
        break
    elif cnt==0:
        print(0)
        break
    
    #1년후 변화 반영하기
    for i in range(n):
        for j in range(m):
            if sea[i][j]>0:
                sea[i][j]=max(0, sea[i][j]-(visited[i][j]-1))
    year+=1
```

### 4. 노드사이의 거리

```python
import sys
from collections import deque
from io import StringIO
sys.stdin=StringIO("""\
4 2
2 1 2
4 3 2
1 4 3
1 2
3 2
""")

input=sys.stdin.readline

n,m = map(int, input().split())
graph=[[] for _ in range(n+1)]
for _ in range(n-1):
    a, b, d = map(int, input().split())
    graph[a].append((b,d))
    graph[b].append((a,d))
    
def bfs(visited, x,y):
    total_d=0
    q=deque()
    q.append((x, total_d))
    visited[x]=True
    while q:
        curr, total_d = q.popleft()
        for node,d in graph[curr]:
            if node==y:
                return total_d+d
            elif node!=y and not visited[node]:
                q.append((node, total_d+d))
                visited[node]=True
        
            

for _ in range(m):
    x, y =map(int, input().split())
    visited=[False]*(n+1)
    print(bfs(visited,x,y))
```

### 5. 떡볶이 떡 만들기

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
4 6
19 15 10 17
""")
input=sys.stdin.readline

n,m = map(int, input().split())
high = list(map(int, input().split()))

start, end = 0, max(high)

result=None
while(start<=end):
    ready=0
    mid = (start+end)//2
    #자른 후의 총 떡 합 구하기
    for x in high:
        if x >mid:
            ready+=(x-mid)
    #요청보다 작은 경우 end 조절
    if ready<m:
        end = mid-1
    #요청 이상인 경우 일단 result에 넣고 start 조절
    else:
        result=mid
        start=mid+1
print(result)
```

### 6. 정렬된 배열에서 특정 수의 개수 구하기

```python
import sys
from io import StringIO
from bisect import bisect_left, bisect_right

sys.stdin=StringIO("""\
7 2
1 1 2 2 2 2 3
""")
input=sys.stdin.readline

n, x = map(int, input().split())
array = list(map(int, input().split()))

min = bisect_left(array, x)
max = bisect_right(array, x)

print(max-min)
```
