### 1. 적록색약 : https://www.acmicpc.net/problem/10026

```python
from collections import deque
import sys
from io import StringIO

sys.stdin=StringIO("""\
5
RRRBB
GGBBB
BBBRR
BBRRR
RRRRR
""")

input = sys.stdin.readline

n = int(input())
graph=[]
graph2=[]
for i in range(n):
    pic = input()
    graph.append(list(pic.strip()))
    graph2.append(list(pic.replace('G','R').strip()))


def bfs(m, i,j):
    q=deque()
    q.append((i,j))
    col = m[i][j]
    m[i][j]=None
    while q:
        x, y = q.popleft()
        for dx, dy in [(-1,0),(1,0),(0,-1),(0,1)]:
            nx, ny = x+dx, y+dy
            if 0<=nx<n and 0<=ny<n and m[nx][ny]==col:
                q.append((nx, ny))
                m[nx][ny]=None

cnt=0
cnt2=0
for i in range(n):
    for j in range(n):
        if graph[i][j]:
            bfs(graph, i,j)
            cnt+=1
        if graph2[i][j]:
            bfs(graph2,i,j)
            cnt2+=1

print(cnt, cnt2)


#적록색약 : 빨강==초록
```

### 2. 이분그래프 : https://www.acmicpc.net/problem/1707

```python
import sys
from io import StringIO
from collections import deque
sys.stdin=StringIO("""\
2
3 2
1 3
2 3
4 3
1 2
2 3
3 4
""")

input=sys.stdin.readline

k = int(input())
for _ in range(k):
    V, E = map(int, input().split())  #정점개수 간선개수
    graph=[[]for _ in range(V+1)]
    color=[0]*(V+1)  #1 or -1로 색칠

    for _ in range(E):
        u, v = map(int, input().split())
        graph[u].append(v)
        graph[v].append(u)
    
    ok=True
    for stt in range(1, V+1):
        if color[stt]!=0:  #색칠 안된 정점에 대해서만 검사 
            continue
        color[stt]=1
        q=deque()
        q.append(stt)
        while q and ok:  #연결된 모든 정점들의 색은 다른색이어야함  
            x=q.popleft()
            for nxt in graph[x]:
                if color[nxt]==0: #아직 안칠해진 경우
                    color[nxt]=-color[x]
                    q.append(nxt)
                elif color[nxt]==color[x]: #같은색인 경우
                    ok=False
                    break
        if not ok:
            break
        

    print('YES' if ok else 'NO')
```
