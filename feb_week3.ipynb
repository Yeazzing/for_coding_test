### 1. 개미 전사

```python
import sys
from io import StringIO
sys.stdin=StringIO("""\
4
1 3 1 5
""")
input=sys.stdin.readline

n = int(input())
food = list(map(int, input().split()))

#보텀업 다이나믹 프로그래밍
d = [0]*n
d[0]=food[0]
d[1]=max(food[0], food[1])

for i in range(2, n):
    d[i] = max(d[i-1], d[i-2]+food[i]) #점화식
    
print(d[n-1])
```

### 2. 1로 만들기

```python
sys.stdin=StringIO("""\
26
""")
input=sys.stdin.readline

x = int(input())

d = [0]*(x+1)

for i in range(2, x+1):
    d[i] = d[i-1]+1  #현재의 수에서 1을 빼는 경우
    if i%5==0:  #5로 나누어 떨어지는 경우
        d[i] = min(d[i], d[i//5]+1)
    if i%3==0:  #3으로 나누어 떨어지는 경우
        d[i] = min(d[i], d[i//3]+1)
    if i%2==0:  #2로 나누어 떨어지는 경우
        d[i] = min(d[i], d[i//2]+1)
        

print(d[x])
```

### 3. 효율적인 화폐 구성

```python
sys.stdin=StringIO("""\
3 4
3
5
7
""")
input=sys.stdin.readline

n,m = map(int, input().split())
coin=[]
for i in range(n):
    x = int(input())
    coin.append(x)
    
d = [10001]*(m+1)
d[0]=0

for i in range(n):  #각 화폐 단위
    for j in range(coin[i],m+1):  #만들어야하는금액
        if d[j-coin[i]]!=10001:  #만들어야하는금액-화폐금액 만드는 방법이 존재하는 경우
            d[j]=min(d[j], d[j-coin[i]]+1)  #더 작은 값으로 갱신
            

if d[m]==10001:
    print(-1)
else:
    print(d[m])
```

### 4. 금광

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
2
3 4
1 3 3 2 2 1 4 1 0 6 4 7
4 4
1 3 1 5 2 2 4 1 5 0 2 3 0 6 1 2
""")
input=sys.stdin.readline

t = int(input())
for _ in range(t):
    n, m = map(int, input().split())
    gold= list(map(int, input().split()))
    
    #dp테이블 초기화
    dp=[]
    index=0
    for i in range(n):
        dp.append(gold[index:index+m])
        index+=m
    
    #다이나믹프로그래밍 진행 : 왼쪽열부터 각 칸을 채워나가기
    for j in range(1,m):
        for i in range(n):
            #왼쪽위
            if i==0:
                leftup=0
            else:
                leftup=dp[i-1][j-1]
            #왼쪽 아래
            if i==n-1:
                leftdown=0
            else:
                leftdown=dp[i+1][j-1]
            #왼쪽
            left=dp[i][j-1]
            
            dp[i][j]=dp[i][j]+max(leftup,leftdown,left)
    result=0
    for i in range(n):
        result=max(result,dp[i][m-1])
    print(result)
```

### 5. 병사 배치하기

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
7
15 11 4 8 5 2 4
""")
input=sys.stdin.readline

n = int(input())
array=list(map(int, input().split()))
array.reverse()

d = [1]*n
#가장 긴 증가하는 부분 수열 알고리즘
for i in range(1, n):
    for j in range(0, i):  #i앞 원소들을 돌면서
        if array[j]<array[i]: #i값보다 작으면
            d[i] = max(d[i], d[j]+1) #max(i의 현재 증가개수 값, 원소j의증가개수값+1)
            
print(n-max(d)) #전체-증가개수 = 열외개수
```

### 6. 설탕 배달

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
11
""")
input=sys.stdin.readline

n = int(input())
d = [5001]*(n+1)
d[0]=0

for i in range(3,n+1):
    if d[i-3]!=5001:
        d[i]=min(d[i],d[i-3]+1)
    if i-5>=0 and d[i-5]!=5001:
        d[i]=min(d[i],d[i-5]+1)
        
if d[n]==5001:
    print(-1)
else:
    print(d[n])
```

### 7. 1, 2, 3 더하기

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
3
4
7
10
""")
input=sys.stdin.readline

t = int(input())
for _ in range(t):
    n = int(input())
    
    d=[0]*(n+1)
    d[0]=1
    
    for i in range(n+1):
        if i-1>=0:
            d[i]+=d[i-1]
        if i-2>=0:
            d[i]+=d[i-2]
        if i-3>=0:
            d[i]+=d[i-3]
            
    print(d[n])
```

### 8. RGB 거리

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
8
71 39 44
32 83 55
51 37 63
89 29 100
83 58 11
65 13 15
47 25 29
60 66 19
""")
input=sys.stdin.readline

n = int(input())
g=[]
for _ in range(n):
    price = list(map(int, input().split()))
    g.append(price)
    

for i in range(1,n):
    for j in range(3):
        g[i][j] =min(g[i-1][j-1]+g[i][j],g[i-1][(j+1)%3]+g[i][j])

answer=1000*n
for p in g[n-1]:
    answer=min(answer,p)
print(answer)
```

### 9. 전보

```python
import sys
from io import StringIO
import heapq

sys.stdin=StringIO("""\
3 2 1
1 2 4
1 3 2
""")
input=sys.stdin.readline

INF = int(1e9)

n,m,c = map(int, input().split())
graph =[[]for  i in range(n+1)]
distance=[INF]*(n+1)

for _ in range(m):
    x,y,z = map(int, input().split())
    graph[x].append((y,z))             

def distra(start):
    q=[]
    heapq.heappush(q, (0, start))
    distance[start]=0
    while q:
        dist, now=heapq.heappop(q)
        if distance[now]<dist:
            continue
        for i in graph[now]:
            cost=dist+i[1]
            if cost<distance[i[0]]:
                distance[i[0]]=cost
                heapq.heappush(q, (cost, i[0]))
                
distra(c)
city=0
max_distance=0
for d in distance:
	if d != INF:
		city+=1
		max_distance=max(max_distance, d)

print(city-1, max_distance)
```

### 10. 미래 도시

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
5 7
1 2
1 3
1 4
2 4
3 4
3 5
4 5
4 5
""")
input=sys.stdin.readline
INF=int(1e9)

n, m =map(int, input().split())
graph=[[INF]*(n+1) for _ in range(n+1)]
for _ in range(m):
    x,y=map(int, input().split())
    graph[x][y]=1
    graph[y][x]=1
x,k = map(int, input().split())

for i in range(1,n+1):
    for j in range(1,n+1):
        if i==j:
            graph[i][j]=0

for z in range(1, n+1):
    for a in range(1,n+1):
        for b in range(1,n+1):
            graph[a][b]=min(graph[a][b], graph[a][z]+graph[z][b])

if graph[1][x]==INF:
    print(-1)
else:
    print(graph[1][k]+graph[k][x])
```

### 11. 최단경로

```python
import sys
from io import StringIO
import heapq

sys.stdin=StringIO("""\
5 6
1
5 1 1
1 2 2
1 3 3
2 3 4
2 4 5
3 4 6
""")
input=sys.stdin.readline

v,e=map(int, input().split())
INF=int(1e9)
graph=[[] for _ in range(v+1)]
distance=[INF]*(v+1)
start=int(input())

for _ in range(e):
    u,vv,w=map(int, input().split())
    graph[u].append((vv,w))
    
q=[]
distance[start]=0
heapq.heappush(q, (0,start))
while q:
    d, now = heapq.heappop(q)
    if d > distance[now]:
        continue
    for nxt in graph[now]:
        nowd=d+nxt[1]
        if nowd < distance[nxt[0]]:
            distance[nxt[0]]=nowd
            heapq.heappush(q,(nowd,nxt[0]))
            
for i in range(1,v+1):
    if distance[i]==INF:
        print('INF')
    else:
        print(distance[i])
```

### 12.운동

```python
import sys
from io import StringIO
import heapq

sys.stdin=StringIO("""\
3 4
1 2 1
3 2 1
1 3 5
2 3 2
""")
input=sys.stdin.readline

v,e = map(int, input().split())
INF=int(1e9)
graph=[[]for _ in range(v+1)]
dist=[[INF]*(v+1) for _ in range(v+1)]
q=[]

for _ in range(e):
    a,b,c= map(int, input().split())
    graph[a].append((b,c))
    dist[a][b]=c
    heapq.heappush(q, (c,a,b))
    
while q:
    d, x,y = heapq.heappop(q)
    if x==y:
        print(d)
        break
    if dist[x][y] < d:
        continue
    for ny, nd in graph[y]:
        if d+nd < dist[x][ny]:
            dist[x][ny]=d+nd
            heapq.heappush(q, (d+nd,x,ny))
else:
    print(-1)
```
