### 1. 민트 초코 우유

```python
import sys
from io import StringIO
from collections import deque

sys.stdin=StringIO("""\
4 2
TTCC
TTTM
CCMM
CMMM
1 3 3 3
2 23 16 8
12 6 7 8
12 8 3 5
""")
input=sys.stdin.readline

n, t = map(int, input().split())
food = []
for _ in range(n):
    food.append(list(input().strip()))
like = []
for _ in range(n):
    like.append(list(map(int, input().split())))

dx = [-1, 1, 0,0]
dy = [0,0, -1, 1]

daily = ["CMT", "CT", "MT", "CM", "M", "C", "T"]
    
def morning():
    for i in range(n):
        for j in range(n):
            like[i][j]+=1

def lunch():
    visited=[[False]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            group=[]
            if visited[i][j]==False:
                group.append((like[i][j], -i, -j))
                q = deque()
                q.append((i,j))
                visited[i][j]=True
                while q:
                    x, y = q.popleft()
                    for d in range(4):
                        nx, ny = x+dx[d], y+dy[d]
                        if 0<=nx<n and 0<=ny<n and visited[nx][ny]==False and food[i][j]==food[nx][ny]:
                            visited[nx][ny]=True
                            q.append((nx,ny))
                            group.append((like[nx][ny], -nx, -ny))
                group.sort(reverse=True)
                for k in range(len(group)):
                    if k==0:
                        like[-group[k][1]][-group[k][2]]+=(len(group)-1)
                        num_word = len(food[-group[k][1]][-group[k][2]])
                        rep.append((-num_word, like[-group[k][1]][-group[k][2]], group[k][1], group[k][2]))
                    else:
                        like[-group[k][1]][-group[k][2]]-=1

def dinner(r, c):
    b = like[r][c]
    x = b-1
    like[r][c]=1
    d = b%4
    while True:
        nr, nc = r+dx[d], c+dy[d]
        if 0<=nr<n and 0<=nc<n and x>0:
            if food[r][c]!=food[nr][nc]:
                y = like[nr][nc]
                if x>y:  #강한 전파
                    food[nr][nc]=food[r][c]
                    x-=(y+1)
                    like[nr][nc]+=1
                    attack[nr][nc]=True
                    if x==0:
                        break
                elif x<=y:  #약한 전파
                    food[nr][nc]+=food[r][c]
                    food[nr][nc] = ''.join(sorted(set(food[nr][nc])))
                    like[nr][nc]+=x
                    attack[nr][nc]=True
                    x=0
                    break
            r, c = nr, nc
        else:
            break
        
        
for day in range(t):
    #아침
    rep = []
    attack = [[False]*n for _ in range(n)]
    num_daily=[0]*len(daily)
    morning()
    #점심 - 그룹형성
    lunch()
    #저녁 
    rep.sort(reverse=True) #(-글자수, like수, -행, -열)
    for (_,_,r, c) in rep:  
        if attack[-r][-c]==False:
            dinner(-r,-c)

    for i in range(n):
        for j in range(n):
            for k in range(len(daily)):
                if food[i][j]==daily[k]:
                    num_daily[k]+=like[i][j]
    print(*num_daily)
```

### 2. Bitmask

```python
#and 연산자
print(5 & 6) #101 & 110 -> 100

#or 연산자
print(5 | 3) #101 | 011 -> 111

#xor 연산자 : 두 비트가 다른 값이면 1, 같은 값이면 0
print(5^3) #101^011 -> 110

#left shift 연산자 : 모든 비트를 b번 왼쪽으로 밈 -> 값이 2배가 됨
# 5 = 0000101
print(5<<1) #0001010
print(5<<2) #0010100

#right shift 연산자 : 반대 -> 값이 2로 나눈 몫이 됨
print(13>>1)
print(13>>2)
```

```python
#Bitmask
#정수값 -> 32bit 
#32인 배열로 각 수들이 남아있는지 관리 

#                                     x         상황
# 00000000000000000000000000000000   (0) : 아무 수도 없는 경우
# 00000000000000000000000000001010  (10) : 1, 3만 있는 경우 
# 00000000000000000000000000100111  (39) : 0, 1, 2, 5만 있는 경우 
# 00000000000000000000000001000000  (64) : 6만 있는 경우 
x=39  #00000000000000000000000000100111
#수 k가 존재하는지 여부 확인 
k=2
print((x>>k)&1)

#수 k를 새롭게 추가
print(x+(1<<k))  #00000000000000000000000000110111

#수 k를 있다면 제거 없다면 추가
print(x^(1<<k))  #00000000000000000000000000100111

#오른쪽 n개의 bit를 전부 1로 
n=5
print((1<<n)-1)  #00000000000000000000000000011111
```

```python
#bit 계산 문제
import sys
from io import StringIO

sys.stdin=StringIO("""\
7
add 1
delete 1
print 1
toggle 1
print 1
clear
print 1
""")
input=sys.stdin.readline
result =0

q = int(input())
for _ in range(q):
    line = input().split()
    order=line[0]
    
    if order == 'clear':
        result = 0
        continue
    
    x = int(line[1])
    if str(order)=='add':
        result|=(1<<x)
    elif str(order)=='delete':
        result&=~(1<<x)
    elif str(order)=='print':
        print((result>>x)&1)
    elif str(order)=='toggle':
        result^=(1<<x)
```

### 3. 겹치지 않는 쌍의 수

```python
import sys
from io import StringIO

sys.stdin=StringIO("""\
4
2 1 30
3 5 2 9
3 1 3 5
3 4 30 5
""")
input=sys.stdin.readline

n = int(input())
groups=[]
for _ in range(n):
    group = list(map(int, input().split()))
    g = 0
    for i in range(1, group[0]+1):
        k = group[i]
        g|=(1<<k)
    groups.append(g)

answer=0   
for i in range(n-1):
    for j in range(i+1, n):
        if groups[i]&groups[j]==0:
            answer+=1
print(answer)
```

### 4. 아기 상어

```python
import sys
from io import StringIO
from collections import deque

sys.stdin=StringIO("""\
6
1 1 1 1 1 1
2 2 6 2 2 3
2 2 5 2 2 3
2 2 2 4 6 3
0 0 0 0 0 6
0 0 0 0 0 9
""")
input=sys.stdin.readline

n = int(input())
fish=[]
for _ in range(n):
    fish.append(list(map(int, input().split())))

shark=2
time=0
eat=0

def find(i, j):
    visited=[[False]*n for _ in range(n)]
    can = []
    visited[i][j]=True
    q=deque([(0,i,j)])
    while q:
        d, x, y = q.popleft()
        if 0<fish[x][y]<shark:
            can.append((d,x,y))
        elif 9>fish[x][y]>shark:
            continue
        for dx, dy in [(-1,0), (1,0),(0,-1),(0,1)]:
            nx, ny = x+dx, y+dy
            if 0<=nx<n and 0<=ny<n and visited[nx][ny]==False:
                q.append((d+1, nx,ny))
                visited[nx][ny]=True
    return can

for i in range(n):
    for j in range(n):
        if fish[i][j]==9:
            r, c = i, j
            break
        
while True:
    #물고기 찾기
    can = find(r, c)
    if can:
        fish[r][c]=0  #상어 있던 자리 0으로 
        can.sort()
        d, r, c = can[0]  #현재 위치 변경 (물고기 먹음)
        time +=d
        fish[r][c]=0  #현재 위치 0으로
        eat+=1
        if eat==shark:
            shark+=1
            eat=0
    else:
        print(time)
        break
```
