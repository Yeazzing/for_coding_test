### 1. 민트 초코 우유

```python
import sys
from io import StringIO
from collections import deque

sys.stdin=StringIO("""\
4 2
TTCC
TTTM
CCMM
CMMM
1 3 3 3
2 23 16 8
12 6 7 8
12 8 3 5
""")
input=sys.stdin.readline

n, t = map(int, input().split())
food = []
for _ in range(n):
    food.append(list(input().strip()))
like = []
for _ in range(n):
    like.append(list(map(int, input().split())))

dx = [-1, 1, 0,0]
dy = [0,0, -1, 1]

daily = ["CMT", "CT", "MT", "CM", "M", "C", "T"]
    
def morning():
    for i in range(n):
        for j in range(n):
            like[i][j]+=1

def lunch():
    visited=[[False]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            group=[]
            if visited[i][j]==False:
                group.append((like[i][j], -i, -j))
                q = deque()
                q.append((i,j))
                visited[i][j]=True
                while q:
                    x, y = q.popleft()
                    for d in range(4):
                        nx, ny = x+dx[d], y+dy[d]
                        if 0<=nx<n and 0<=ny<n and visited[nx][ny]==False and food[i][j]==food[nx][ny]:
                            visited[nx][ny]=True
                            q.append((nx,ny))
                            group.append((like[nx][ny], -nx, -ny))
                group.sort(reverse=True)
                for k in range(len(group)):
                    if k==0:
                        like[-group[k][1]][-group[k][2]]+=(len(group)-1)
                        num_word = len(food[-group[k][1]][-group[k][2]])
                        rep.append((-num_word, like[-group[k][1]][-group[k][2]], group[k][1], group[k][2]))
                    else:
                        like[-group[k][1]][-group[k][2]]-=1

def dinner(r, c):
    b = like[r][c]
    x = b-1
    like[r][c]=1
    d = b%4
    while True:
        nr, nc = r+dx[d], c+dy[d]
        if 0<=nr<n and 0<=nc<n and x>0:
            if food[r][c]!=food[nr][nc]:
                y = like[nr][nc]
                if x>y:  #강한 전파
                    food[nr][nc]=food[r][c]
                    x-=(y+1)
                    like[nr][nc]+=1
                    attack[nr][nc]=True
                    if x==0:
                        break
                elif x<=y:  #약한 전파
                    food[nr][nc]+=food[r][c]
                    food[nr][nc] = ''.join(sorted(set(food[nr][nc])))
                    like[nr][nc]+=x
                    attack[nr][nc]=True
                    x=0
                    break
            r, c = nr, nc
        else:
            break
        
        
for day in range(t):
    #아침
    rep = []
    attack = [[False]*n for _ in range(n)]
    num_daily=[0]*len(daily)
    morning()
    #점심 - 그룹형성
    lunch()
    #저녁 
    rep.sort(reverse=True) #(-글자수, like수, -행, -열)
    for (_,_,r, c) in rep:  
        if attack[-r][-c]==False:
            dinner(-r,-c)

    for i in range(n):
        for j in range(n):
            for k in range(len(daily)):
                if food[i][j]==daily[k]:
                    num_daily[k]+=like[i][j]
    print(*num_daily)
```
