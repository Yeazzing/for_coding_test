### 1. 소수 구하기

```python
import math

m, n = map(int, input().split())

nums = [False]*(n+1)

for i in range(2, int(math.sqrt(n)+1)):
    if nums[i]==False:
        j=2
        while i*j<=n:
            nums[i*j]=True
            j+=1

for i in range(2, len(nums)):
    if m<=i<=n and nums[i]==False:
        print(i)
```

### 2. 베르트랑 공준

```python
import math
import sys
from io import StringIO

sys.stdin=StringIO("""\
1
10
13
100
1000
10000
100000
0
""")
input=sys.stdin.readline

nums = [False]*(123456*2+1)
for i in range(2, int(math.sqrt(123456*2)+1)):
    if nums[i]==False:
        j=2
        while i*j<=2*123456:
            nums[i*j]=True
            j+=1

while True:
    n=int(input())
    if n==0:
        break

    answer=0
    for i in range(n+1, 2*n+1):
        if nums[i]==False:
            answer+=1
    print(answer)
```

### 3. 골드바흐의 추측

```python
import math
import sys
from io import StringIO

sys.stdin=StringIO("""\
3
8
10
16
""")
input=sys.stdin.readline
t = int(input())

nums=[False]*(10000+1)
for i in range(2, int(math.sqrt(10000)+1)):
    if nums[i]==False:
        j=2
        while i*j<=10000:
            nums[i*j]=True
            j+=1

for _ in range(t):
    n = int(input())
    for i in range(int(n//2), 1, -1):
        j = n-i
        if nums[i]==False and nums[j]==False:
            print(i,j)
            break
```

### 4. 크로스 컨트리

```python
import math
import sys
from io import StringIO

sys.stdin=StringIO("""\
2
15
1 2 3 3 1 3 2 4 1 1 3 1 3 3 1
18
1 2 3 1 2 3 1 2 3 3 3 3 2 2 2 1 1 1
""")
input=sys.stdin.readline

t = int(input())
for _ in range(t):
    n = int(input())
    team = list(map(int, input().split()))
    result = [[]for _ in range(max(team))]
    score=1
    for i in range(len(team)):
        if team.count(team[i])<6:
            continue
        result[team[i]-1].append(score)
        score+=1
        
    answer=1e9
    for j in range(len(result)):
        if len(result[j])<6:
            continue
        team_score = sum(result[j][0:4])
        if team_score<answer:
            ans_t = j
            answer=team_score
        elif team_score==answer:
            if result[ans_t][4] > result[j][4]:
                ans_t = j
            else:
                continue
    print(ans_t+1)
#팀 점 수 : 상위 네명의 점수 합
#가장 낮은 점수 팀이 우승
#6명이 참가하지 못한 팀은 제외
#동점인 경우 다섯번째 주자가 가장 빨리 들어온 팀이 우승
```
